```python
#create a dictionary feature name : number of nan
total = []
for val in df.isnull().sum():
    total.append(val)
nan = { k : v for k,v in zip(df.columns, total)}
nan
```

```python
missing_percentages = df.isna().mean() * 100
result_df = pd.DataFrame({'column_name': missing_percentages.index, 
                          'Number of missing values': df.isna().sum(),
                          'percent_missing': missing_percentages.values})
print(result_df)
```

```python
#Get the columns that have more than 30% missing values
cols_with_more_than_30percent_missing = result_df[result_df['percent_missing'] > 30]['column_name']

#Print the result
print(cols_with_more_than_30percent_missing)
```

```python
df.drop(columns=cols_with_more_than_30percent_missing, inplace=True)
df
```

```python
# missing values per row (selected column threshold: 30%)
row_list = []
for i in range(df.shape[0]):
    n_miss = df.iloc[i].isnull().sum()
    perc = n_miss / df.shape[1] * 100
    # print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
    if perc >= 30:
        row_list.append(i)
```

```python
row_list
```

```python
df.drop(row_list, inplace=True)
df
```

# Filling missing values by mode of that column

```python
# Iterate over each column in the DataFrame
for column in df.columns:
    # Fill missing values with mode value of that column
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x = df.iloc[:, :-1]  #features
y = df.iloc[:, -1]   #target variable
```

```python
y 
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100)
```

```python
rf_classifier.fit(x, y)
```

```python
importances = rf_classifier.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
print("Feature ranking:")
for f in range(x.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
plt.figure()
plt.title("Feature importances")
plt.barh(x.columns, importances)
plt.xlabel("Feature index (descending order)")
plt.ylabel("Feature importance score")
plt.show()
```

```python
plt.figure(figsize=(8,18)) # set plot size
plt.barh(x.columns, rf_classifier.feature_importances_) # plot feature importances
plt.xlabel('Importance') # set x-axis label
plt.title('Feature Importances') # set plot title
plt.show()
```
