```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
pd.options.display.max_columns = 200
df= pd.read_csv('Myocardial infarction complications.csv', index_col=0)
df
```

```python
#create a dictionary feature name : number of nan

total = []
for val in df.isnull().sum():
    total.append(val)
nan = { k : v for k,v in zip(df.columns, total)}
nan
```

# Handling Missing Values

```python
missing_percentages = df.isna().mean() * 100
result_df = pd.DataFrame({'column_name': missing_percentages.index, 
                          'Number of missing values': df.isna().sum(),
                          'percent_missing': missing_percentages.values})
print(result_df)
```

```python
#Get the columns that have more than 30% missing values

cols_with_more_than_30percent_missing = result_df[result_df['percent_missing'] > 30]['column_name']

#Print the result

print(cols_with_more_than_30percent_missing)
```

```python
df.drop(columns=cols_with_more_than_30percent_missing, inplace=True)
df
```

```python
#Missing values per row (selected column threshold: 30%)

row_list = []
for i in range(df.shape[0]):
    n_miss = df.iloc[i].isnull().sum()
    perc = n_miss / df.shape[1] * 100
    # print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
    if perc >= 30:
        row_list.append(i)
```

```python
row_list
```

```python
df.drop(row_list, inplace=True)
df
```

```python
x = df.iloc[:, :-1]  #features
y = df.iloc[:, -1]   #target variable
```

```python
#Create list of binary variables

binary_list = []
for i in range(len(x.iloc[0])):
    if x.iloc[:,i].max() == 1:
        binary_list.append(i)

binary_list
```

```python
#Create an empty list to store the index numbers and names of continuous variables

continuous_list = []

#Iterate over each column in the DataFrame

for i, col in enumerate(df.columns):
    
    #Check if the column contains numeric values
    
    if pd.api.types.is_numeric_dtype(df[col]):
        
        #Check if the column has more than 10 unique values, indicating it is likely a continuous variable
        
        if df[col].nunique() > 10:
            
            #Add the column index and name to the list of continuous variables
            
            continuous_list.append((i))
            
continuous_list
```

```python
#Create list of all variables

listt = [*range(115)]
```

```python
def Diff(li1, li2):
    return list(set(li1) - set(li2)) + list(set(li2) - set(li1))
```

```python
#create list of categorical variables 

binary_categorical = Diff(listt, continuous_list)
categorical_list = Diff(binary_categorical,binary_list)
categorical_list
```

```python
#convert binary to boolean

for i in range(len(x.columns[binary_list])):
    x[x.columns[binary_list][i]] = x[x.columns[binary_list][i]].astype('bool')
x[x.columns[binary_list]]
x_binary = x[x.columns[binary_list]]
binary_name = x_binary.columns
binary_name
```

```python
#convert continuous to float

for i in range(len(x.columns[continuous_list])):
    x[x.columns[continuous_list][i]] = x[x.columns[continuous_list][i]].astype('float')
x[x.columns[continuous_list]]
x_continuous = x[x.columns[continuous_list]]
cont_name = x_continuous.columns
cont_name
```

```python
#convert categorical to integer

for i in range(len(x.columns[categorical_list])):
    x[x.columns[categorical_list][i]] = x[x.columns[categorical_list][i]].astype('float')
x[x.columns[categorical_list]]
x_categorical = x[x.columns[categorical_list]]
catg_name = x_categorical.columns
catg_name
```

# Data imputation
    For the remaining rows and columns with missing value ratio of less than 30%, I apply basic    data imputation.

    For binary and continuous variables, I apply k-nearest neighbor imputation. As the name         suggests, the ‘k-nearest neighbor’ imputation strategy looks at nearest neighbors to determine the imputed value.

```python
#impute binary variables 

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
imputer=imputer.fit(x_binary)

#convert to array

x_binary = imputer.transform(x_binary)
x_binary = pd.DataFrame(x_binary)
x_binary.columns = binary_name

#Check null values

print("Total sum of all missing values in X_binary:", x_binary.isnull().sum(axis = 0).sum())
```

```python
#impute continuous variables

imputer = KNNImputer(n_neighbors=5)
imputer=imputer.fit(x_continuous)

x_continuous = imputer.transform(x_continuous)
x_continuous = pd.DataFrame(x_continuous)
x_continuous.columns = cont_name
print("Total sum of all missing values in X_continuous:", x_continuous.isnull().sum(axis = 0).sum())
```

    For categorical variables, I apply the most frequent imputation strategy. This prevents      
    categorical values from having fractional values and the missing values will be replaced by   the mode.

```python
#impute categorical variables

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer=imputer.fit(x_categorical)

x_categorical = imputer.transform(x_categorical)
x_categorical = pd.DataFrame(x_categorical)
x_categorical.columns = catg_name

print("Total sum of all missing values in X_categorical:",x_categorical.isnull().sum(axis = 0).sum())
```

# One-hot Encoding

    After imputing missing values, our categorical variables still had the problem of being ordered. I convert these features into numbers using one-hot-encoding. With one-hot-encoding, each  categorical level becomes a separate feature in the dataset with binary values (1 or 0). This allows machine learning algorithms to read categorical data in a better way.

```python
#Returns columns of an array

def column(matrix, i):
    return [row[i] for row in matrix]
```

```python
#Encode categorical variables

x_cat = x_categorical.to_numpy()
encoded_dataframe = pd.get_dummies(column(x_cat,0))

for i in range(1,len(x_categorical.iloc[0])):
    encoded = pd.get_dummies(column(x_cat,i))
    encoded_dataframe = pd.concat([encoded_dataframe, encoded], axis=1)
    
categorical_names = [*np.arange(101)]
encoded_dataframe.columns = categorical_names
encoded_dataframe
```

#  Normalization
    The continuous variables has different scales, which is a problem. This would cause continuous variables with large absolute values to affect the algorithm more.

    To avoid this problem, I apply centering. I subtract the mean and divide by the standard    
    deviation of each variable. This way, each continuous variable has zero mean and one unit of  standard deviation.

```python
# standardize continuous variables
for i in range(len(x_continuous.iloc[0])):
    x_continuous.iloc[:, i] = (x_continuous.iloc[:, i] - x_continuous.iloc[:, i].mean())/x_continuous.iloc[:, i].std()

#combine datasets

x_all_combined = pd.concat([x_continuous, encoded_dataframe, x_binary], axis=1)
x= x_all_combined
```

# Train-test Split

    I partition the data into training, cross-validation and test sets. To ensure equally balanced classes that best represent the population, I apply stratified sampling during the            train-valid-test split.

    Using the training set, I fit model parameters which minimize the error between the actual    labels and the predicted labels.

    Then, for each model family, I select the best model on the cross-validation set, using       10-fold cross validation.

    Finally, I compute the generalization error using the test set.

```python
y.value_counts()
```

# Balancing Dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=143)
```

```python
y_train.shape, y_test.shape
```

```python
#Apply SMOTE to the training set

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, sampling_strategy='auto')
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
x_train_resampled.shape, y_train_resampled.shape
```

```python
sns.countplot(y_train_resampled)
```

# Decision Tree


```python
#Decision tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

tree = DecisionTreeClassifier(random_state = 42)
tree.fit(x_train_resampled, y_train_resampled)
y_pred = tree.predict(x_test)

#Calculate the F1 score for each class using the "macro" averaging method
f1_macro = f1_score(y_test, y_pred, average='macro')

#Calculate the F1 score for each class using the "micro" averaging method
f1_micro = f1_score(y_test, y_pred, average='micro')

#Calculate the F1 score for each class using the "weighted" averaging method
f1_weighted = f1_score(y_test, y_pred, average='weighted')

print("f1_macr0:",f1_macro)
print("f1_micro:",f1_micro)
print("f1_weighted:",f1_weighted)

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score

y_train_pred = tree.predict(x_train_resampled)

# Calculate the training and testing accuracy
train_acc = accuracy_score(y_train_resampled, y_train_pred)
test_acc = accuracy_score(y_test, y_pred)

print('Training accuracy:', train_acc)
print('Testing accuracy:', test_acc)
```

```python
print(f'Train acc = {tree.score(x_train_resampled, y_train_resampled)}')
print(f'Test acc = {tree.score(x_test, y_test)}')
```

```python
y_test_map = []
for el in y_test:
  if el == 1:
    y_test_map.append(0)
  else:
    y_test_map.append(1)
y_test_map = np.array(y_test_map)
y_test_map

y_pred_map = []
for el in y_pred:
  if el == 1:
    y_pred_map.append(0)
  else:
    y_pred_map.append(1)
y_pred_map = np.array(y_pred_map)

from sklearn import metrics
from sklearn.metrics import roc_curve, auc

false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test_map, y_pred_map)
roc_auc = auc(false_positive_rate1, true_positive_rate1)
plt.figure(figsize = (10, 8))
plt.title('Receiver Operating Characteristic - Decision Tree')
plt.plot(false_positive_rate1, true_positive_rate1,color='darkorange')
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```
