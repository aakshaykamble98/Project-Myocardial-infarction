```python
pd.options.display.max_columns = 130
dataframe= pd.read_csv('Myocardial infarction complications.csv', index_col=0)
dataframe
```

```python
# missing values per column (# selected column threshold: 30%)
column_list = []
for i in range(dataframe.shape[1]):
    n_miss = dataframe.iloc[:,i].isnull().sum()
    perc = n_miss / dataframe.shape[0] * 100
    # print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
    if perc >= 30:
        column_list.append(i)
```

```python
column_list
```

```python
# drop rows with missing values
dataframe_col_name = dataframe.columns
dataframe_col_name_delete= dataframe_col_name[column_list]
dataframe.drop(dataframe_col_name_delete, axis=1, inplace=True)
```

```python
#missing values per row (selected column threshold: 30%)
row_list = []
for i in range(dataframe.shape[0]):
    n_miss = dataframe.iloc[i].isnull().sum()
    perc = n_miss / dataframe.shape[1] * 100
    # print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
    if perc >= 30:
        row_list.append(i)
```

```python
dataframe.drop(row_list,axis=0,inplace=True)
```

```python
X = dataframe.iloc[:, :-1]  #features
y = dataframe.iloc[:, -1]   #target variable
```

```python
# create list of binary variables
binary_list = []
for i in range(len(X.iloc[0])):
    if X.iloc[:,i].max() == 1:
        binary_list.append(i)
```

```python
# Create an empty list to store the index numbers and names of continuous variables
continuous_list = []

# Iterate over each column in the DataFrame
for i, col in enumerate(dataframe.columns):
    
    # Check if the column contains numeric values
    if pd.api.types.is_numeric_dtype(dataframe[col]):
        
        # Check if the column has more than 10 unique values, indicating it is likely a continuous variable
        if dataframe[col].nunique() > 10:
            
            # Add the column index and name to the list of continuous variables
            continuous_list.append((i, col))

# Print the index numbers and names of the continuous variables
for i, var in continuous_list:
    print(i, var)
```
