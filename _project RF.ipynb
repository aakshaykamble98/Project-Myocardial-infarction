```python
data_dup = df.duplicated() .any()
print(data_dup)
```

```python
df = df.fillna(df.mean())
df.isnull().sum()
```

```python
import numpy as np
df['outcome'] = np.where(
    df['LET_IS'] == 0, 0, 1)
df.head()
```

```python
df = df.drop('S_AD_KBRIG', 1)
df = df.drop('D_AD_KBRIG', 1)
df = df.drop('S_AD_ORIT', 1)
df= df.drop('D_AD_ORIT', 1)
```

```python
#column_modes = data.mode(axis = 0)
#print(column_modes)
#data = data.fillna(column_modes.iloc[0])
df = df.fillna(-1)
```

```python
df
```

```python
Y = df['outcome']
X = df.iloc[:,1:107]
plt.figure(figsize=(8,5))
df['outcome'].value_counts().plot(kind='bar', color = ['blue', 'yellow'])
my_xticks = ['Alive', 'Dead']
x = df.outcome.unique()
plt.xticks(x, my_xticks)
plt.xlabel("Outcome")
plt.ylabel("Count")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=100, test_size=0.3)
print(X_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
conf_mat = confusion_matrix(Y_test, Y_pred)
print(conf_mat)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat,display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
pred = clf.predict(X_test)
print(classification_report(Y_test, pred))
```

```python
clf.feature_importances_
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,18)) # set plot size
plt.barh(X.columns, clf.feature_importances_) # plot feature importances
plt.xlabel('Importance') # set x-axis label
plt.title('Feature Importances') # set plot title
plt.show() # show plot
```

```python
sorted_idx = clf.feature_importances_.argsort()
plt.barh(X.columns[sorted_idx][:10], clf.feature_importances_[sorted_idx][:10])
plt.xlabel("Random Forest Feature Importance")
```
