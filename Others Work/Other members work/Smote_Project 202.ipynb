## Splitting the data

```python
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

x = df.iloc[:,:-12]
y = df['LET_IS']
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(x, y)
```

```python
plot_importance(xgb_model, max_num_features=60)
plt.grid(visible=None)
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
pca = PCA()
pcs = pca.fit_transform(x)
pcs
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca = PCA(n_components=71)
pcs = pca.fit_transform(x)
pcs
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
col = []

for i in range(1,72):
    i = str(i)
    name = 'PC' + i
    col.append(name)
col
```

```python
data = pd.DataFrame(pcs, columns=col)
data['LET_IS'] = y.values
data
```

```python
x = data.iloc[:,:-1]
y = data['LET_IS']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=1, sampling_strategy='auto')
xtrain_re, ytrain_re = smote.fit_resample(xtrain, ytrain)
```

### XGBoost model

```python
xgb_model = XGBClassifier()
kfold = KFold(n_splits=10)
xgb_model.fit(xtrain_re, ytrain_re)
ypred = xgb_model.predict(xtest)
ypred
```

c1 = cross_val_score(xgb_model, xtrain_re, ytrain_re, cv=kfold)
c1.mean()*100

```python
print(classification_report(ytest, ypred))
```

```python
a1 = accuracy_score(ytest, ypred)
a1
```
